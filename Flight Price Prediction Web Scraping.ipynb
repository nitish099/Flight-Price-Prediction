```python
# importing libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
# connecting to webdriver
driver = webdriver.Chrome(r"C:\Users\Nitish\Downloads\Compressed\chromedriver.exe")
time.sleep(2)
```

```python
# send get request to the webpage server to get the source code of the page
url = 'https://yatra.com/'
driver.get(url) 
time.sleep(4)
```

```python
# Maximize current window
driver.maximize_window()
```

```python
# clicking on search button option
driver.find_element_by_id("BE_flight_flsearch_btn").click()
```

```python
airline1=[]
source1=[]
des1=[]
dur1=[]
dep1=[]
arvl1=[]
totstops1=[]
price1=[]
```

```python
# extracting airline name tags
try:
    for i in driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']"):
        airline1.append(i.text)
except NoSuchElementException:
    airline1.append("No detail")
    
# extracting source tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p[1]"):
        source1.append(i.text)
except NoSuchElementException:
    source1.append("No detail")
    
# extracting destination tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"):
        des1.append(i.text)
except NoSuchElementException:
    des1.append("No detail")
    
# extracting duration tags
try:
    for i in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']"):
        dur1.append(i.text)
except NoSuchElementException:
    dur1.append("No detail")
    
# extracting departure time tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='i-b pr']"):
        dep1.append(i.text)
except NoSuchElementException:
    dep1.append("No detail")
    
# extracting arrival time tags
try:
    for i in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']"):
        arvl1.append(i.text)
except NoSuchElementException:
    arvl1.append("No detail")
    
# extracting total stops tags
try:
    for i in driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']"):
        totstops1.append(i.text)
except NoSuchElementException:
    totstops1.append("No detail")
    
# extracting price tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='pr i-b mr-2']"):
        price1.append(i.text)
except NoSuchElementException:
    price1.append("No detail")
```

```python
len(price1)
```

```python
flight1=pd.DataFrame({})            # making dataframe
flight1['Airline']= airline1
flight1['Date_of_Journey']=doj1
flight1['Source']= source1
flight1['Destination']= des1
flight1['Duration']= dur1
flight1['Departure Time']= dep1
flight1['Arrival Time']= arvl1
flight1['Total Stops']=totstops1
flight1['Price']= price1
flight1
```

```python
# extracting airline name tags
try:
    for i in driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']"):
        airline15.append(i.text)
except NoSuchElementException:
    airline15.append("No detail")
    
# extracting source tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p[1]"):
        source15.append(i.text)
except NoSuchElementException:
    source15.append("No detail")
    
# extracting destination tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"):
        des15.append(i.text)
except NoSuchElementException:
    des15.append("No detail")
    
# extracting duration tags
try:
    for i in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']"):
        dur15.append(i.text)
except NoSuchElementException:
    dur15.append("No detail")
    
# extracting departure time tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='i-b pr']"):
        dep15.append(i.text)
except NoSuchElementException:
    dep15.append("No detail")
    
# extracting arrival time tags
try:
    for i in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']"):
        arvl15.append(i.text)
except NoSuchElementException:
    arvl15.append("No detail")
    
# extracting total stops tags
try:
    for i in driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']"):
        totstops15.append(i.text)
except NoSuchElementException:
    totstops15.append("No detail")
    
# extracting price tags
try:
    for i in driver.find_elements_by_xpath("//div[@class='pr i-b mr-2']"):
        price15.append(i.text)
except NoSuchElementException:
    price15.append("No detail")
```
